# Use Netflix dataset to visualize the distribution of release years for Movies VS TV Shows using Seaborn's histplot()
## Use different color for Movie Vs TV Show

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('netflix_data.csv')
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
df = df.dropna(subset=['release_year'])
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.histplot(data=df, 
             x='release_year', 
             hue='type',              # Different color for Movie vs TV Show
             multiple='stack',        # Stack histograms by type
             palette='Set2',          # Soft distinct colors
             bins=30,                 # Number of bins
             edgecolor='black')
plt.title('Distribution of Netflix Release Years: Movies vs TV Shows', fontsize=14)
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
