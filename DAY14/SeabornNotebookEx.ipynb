# Seaborn Library of Matplotlib
## Seaborn is a data visualization library in Matplotlib that provides high level interface for creating interactive and more colorful plots with less code.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
palette=sns.color_palette()
print(palette)
```

```python
sns.palplot(palette)
```

```python
palette_green=sns.color_palette('Greens', 15)
sns.palplot(palette_green)
```

```python
data_dict={
    'building':['burz-khalifa','medrika','effil-tower'],
    'building_height':[150,120,100]
    }
data=pd.DataFrame(data_dict)
sns.barplot(x='building', y='building_height',data=data)
plt.show()
```

```python
data_dict={
    'building':['burz-khalifa','medrika','effil-tower','Wutub minar'],
    'category': ['highest', 'medium', 'low', 'medium'],
    'building_height':[150,120,100,110]
    }
bar_colors=['red','green','yellow']
data=pd.DataFrame(data_dict)
sns.barplot(x='building', y='building_height',data=data, hue='category',palette=bar_colors)
plt.show()
```

```python
data_iris=sns.load_dataset('iris')
data_iris
```

```python
sns.barplot(x='species', y='sepal_width',data=data_iris)
plt.show
```

```python
sns.barplot(x='species', y='sepal_width',data=data_iris,palette=bar_colors)
plt.tight_layout()
plt.show
```

```python
sns.violinplot(x='species', y='sepal_width',data=data_iris, hue='species',palette=bar_colors)
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.tight_layout()
plt.show
```

```python
sns.barplot(x='species', y='sepal_width',data=data_iris,hue='sepal_width',palette=bar_colors)
plt.tight_layout()
plt.show
```

```python
sns.violinplot(x='species', y='sepal_width',data=data_iris)
```

```python
sns.violinplot(x='species', y='sepal_width',data=data_iris, hue='species',palette='Set2')
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y', linestyle='--',color='pink',alpha=0.8)
```

```python
sns.stripplot(x='species', y='sepal_width',data=data_iris, hue='species',palette='Set2')
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y', linestyle='--',color='pink',alpha=0.8)
```

```python
sns.pairplot(data=data_iris, hue='species')
plt.show()
```

```python
data_iris=sns.load_dataset('iris')
sns.pairplot(data=data_iris, hue='species')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Darkgrid Theme Example using Iris Data')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Darkgrid Theme Example using Iris Data')
plt.show()
```

```python
sns.set_style('dark')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Darkgrid Theme Example using Iris Data')
plt.show()
```

```python
sns.set_style('ticks')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Darkgrid Theme Example using Iris Data')
plt.show()
```

```python
# can use whitegrid darkgrid, dark, white, ticks
sns.set_style('whitegrid')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Darkgrid Theme Example using Iris Data')
plt.xlabel('Sepal Length in cm')
plt.ylabel('Sepal Width in cm')
plt.xticks([4,5,6,7,8], fontsize=10,color='darkblue')
plt.yticks([1,2,3,4,5,6,7], fontsize=10,color='purple')
sns.despine()
plt.show()
```
