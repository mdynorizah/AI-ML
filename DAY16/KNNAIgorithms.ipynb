# KNN Algorithm in Supervised Machine Learning
# KNeighborsClassifier of sklearn library for this one

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = np.array ([
    [1, 3500],
    [2, 3200],
    [3, 3000],
    [4, 2500],
    [5, 2300],
    [6, 2000],
    [7, 2200],
    [2, 3700],
])
y = ['Unfit','Unfit','Unfit','Fit','Fit','Fit','Fit','Unfit']
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X,y)
```

```python
new_person = [[4.5, 2600]]
pred = model.predict(new_person)
print ("Predicted Result:", pred[0])
```

```python
exercise_hours=float(input('Enter Number of Exercise Hours'))
cal_intake=float(input('Enter number of calories consumed'))
pred_fu = model.predict([[exercise_hours, cal_intake]])
print(f'Exercise Hours: {exercise_hours}, Calories Intake: {cal_intake} â†’ You are likely {pred_fu[0]}')
```

# Download student_grade.csv from GitHub.com
# Predict a students' **grade level (A, B, or C)**
# based on **hours studied** and **assignments completed**
# using the K-Nearest Neighbors algorithm.

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('student_grade.csv')
df
```

```python
X = df[['Hours_Studied','Assignments_Completed']]
y = df['Grade']
```

```python
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X,y)
```

```python
pre_grade = pd.DataFrame([[1.5, 1]], columns=['Hours_Studied', 'Assignments_Completed'])
pred = model.predict(pre_grade)
print("Predicted Grade:", pred[0])
```
