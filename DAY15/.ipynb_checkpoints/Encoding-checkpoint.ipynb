```python
#Label Encoding Basic Example One
from sklearn.preprocessing import LabelEncoder

colors = ['StrawBerry','Apple','Banana','Mango','Orange','Apple','Orange','StrawBerry']


le = LabelEncoder()
# Fit and transform
encoded = le.fit_transform(colors)

print(encoded)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

data = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red']})
# Create encoder
ohe = OneHotEncoder(sparse_output=False)

# Fit and transform
encoded = ohe.fit_transform(data[['Color']])

encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(['Color']))
print(encoded_df)
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

data = pd.DataFrame({'Satisfaction': ['Low', 'Medium', 'East', 'High', 'Low']})
le=LabelEncoder()
data['Satisfaction_Encoded'] = le.fit_transform(data['Satisfaction'])
print(data)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data = pd.DataFrame({'Region': ['North', 'South', 'East', 'West', 'South']})
ohe=OneHotEncoder(sparse_output=False)
encoded=ohe.fit_transform(data[['Region']])
encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(['Region']))
print(encoded_df)
```

# Liner Regression:Finds the best fitting straight line
## Step 1: Collect the Data
## Step 2: Create the Model for LinearRegression
## Step 3: Train the Model fix (X,y)
## Step 4: Predict the Result.  predict([[value_to_predict]])
## Step (optional) 5: Visualize using plot of best fitted

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
## Step1:Collect and Prepare the Data
hours_studied = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])
marks_obtained = np.array([20, 25, 30, 35, 50, 60, 70, 80])
```

```python
## Step2:Create the Model for LinearRegression
model=LinearRegression()
```

```python
## Step3:Train the Model Fix (X,y)
model.fit(X=hours_studied,y=marks_obtained)
```

```python
## Step4:Predict the Result predict ([[value_to_predict]])
predicted_marks=model.predict([[9]])
print('predicted Marks for 9 hours',predicted_marks)
```

```python
hour_study=float(input('Number of hours study to predict marks'))
p_marks=model.predict([[hour_study]])
print(f'predicted Marks for {hour_study} hours is likely {p_marks}')
```
