```python
import matplotlib.pyplot as plt
```

```python
month=['Jan','Feb','Mar','April']
sales=[2000,3000,2500,3600]
```

# plt.plot(x,y,color='color name',linestyle='line_style',linewidth=value,
# marker='marker symbol' label='label name')

```python
plt.plot(month, sales, color='hotpink', linestyle='--', linewidth=2, marker='o', label='Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales per Month')
plt.title('Monthly Sales Trend')
plt.legend()
plt.show()
```

```python
plt.plot(month, sales, color='green', linestyle='--', linewidth=2, marker='o', label='Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales per Month')
plt.title('Monthly Sales Trend')
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
plt.plot(month, sales, color='green', linestyle='--', linewidth=2, marker='*', label='Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales per Month')
plt.title('Monthly Sales Trend')
plt.xlim(0,4)
plt.ylim(2000,5000)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
```

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label='Liner')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3, label='Cubic')
plt.xlabel('X Lines')
plt.ylabel('Y Lines')
plt.title('Simple Lines Plot')
plt.legend(loc='upper left')
plt.show()
```

# Bar Plot Excample
  ## plt.bar(x,y,color='',edgecolor='')

```python
categories=['Electronics','Mobile','Books','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['red','blue','green','purple','orange']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors,edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.title('Categories Wise Sales Report')
plt.show()
```

```python
categories=['Electronics','Mobile','Books','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['red','blue','green','purple','orange']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors,edgecolor='red',label='Sales Info')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.title('Categories Wise Sales Report')
plt.grid(axis='y', linestyle='--', color='green', alpha=0.5)
plt.xticks(rotation=30, fontsize=12, color='magenta')
plt.yticks([20000,40000,60000,80000,100000],['20K','40K','60K','80K','1M'], fontsize=12, color='green')
plt.legend(title='Legend',fontsize=8)
plt.tight_layout()
plt.show()
```

# Scatter Plot
  ## plt.scatter(x,y,s='',c='colorName',marker='makername',alpha=transp,edgecolor='color name',label='label of legend')

```python
import matplotlib.pyplot as plt
height_cm = [150, 155, 160, 165, 170, 175, 180, 185]
weight_kg = [50, 53, 56, 59, 62, 66, 68, 72]
plt.figure(figsize=(8,5))
plt.scatter(height_cm, weight_kg, s=100, c='green', marker='v', alpha=0.9, edgecolor='black', label='Weight-Height Comparison')
plt.title('Scatter Plot: Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.show()
```

```python
height_cm = [150, 155, 160, 165, 170, 175, 180, 185]
weight_kg = [50, 53, 56, 59, 62, 66, 68, 72]
plt.figure(figsize=(8,5))
plt.scatter(height_cm, weight_kg, s=100, c='green', marker='o', alpha=1, edgecolor='black', label='Weight/Height')
plt.title('Scatter Plot Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.xlim(140,190)
plt.ylim(40,90)
plt.xticks(rotation=45, fontsize=12, color='green')
plt.yticks([50, 53, 56, 59, 62, 66, 68, 72],[150, 155, 160, 165, 170, 175, 180, 185],fontsize=12, color='green')
plt.grid(color='lightblue', alpha=0.7, linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
