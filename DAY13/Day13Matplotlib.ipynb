```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
temp=np.random.randint(20,100,size=50)
plt.scatter(x,y,c=temp,cmap='plasma',s=120,edgecolor='black')
plt.colorbar(label='Temperature')
plt.title('Scatter plot Colored by Temperature')
plt.xlabel('X-Random')
plt.ylabel('Y-Random')
plt.show()
```

```python
height_cm = [150, 155, 160, 165, 170, 175, 180, 185]
weight_kg = [50, 53, 56, 59, 62, 66, 68, 72]
age_year = [20, 22, 25, 28, 30, 35, 38, 40]
```

```python
plt.figure(figsize=(8,5))
```

```python
plt.scatter(height_cm,weight_kg,c=age_year,cmap='rainbow',s=120,edgecolor='black')
plt.colorbar(label='Age (years)')
plt.title('Scatter plot: Height Vs Weight (colored by age)')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.grid(color='grey', linestyle='--', linewidth=0.7)
plt.show()
```

# Pie Plot Example
  ## plt.pie(x,labels='',colors='',autoopct='',startagnle='',shadow=,explode=,radius)

```python
depts = ['HR', 'Finance', 'IT', 'Marketing', 'Operations']
expenses = [20000, 25000, 30000, 15000, 10000]
```

```python
plt.figure(figsize=(7,7))
```

```python
plt.pie(expenses,
    labels=depts,
    autopct='%1.0f%%',       
    startangle=140,
    explode=(0, 0.05, 0, 0, 0),
    shadow=True)
plt.title('Department wise Expense Report')
plt.show()
```

# create a pie plot using following Data
## sales_data=('Mobile':45000,'Clothing':30000, 'Groceries':35000,'Home Decor':20000)
### show percentage with two decimal 
### use colormap to give the different color to your pie

```python
Sales_data = {'Mobile': 45000, 'Clothing': 30000, 'Groceries': 35000, 'Home Decor': 20000}
depts = list(Sales_data.keys())
expenses = list(Sales_data.values())
plt.figure(figsize=(7,7))
plt.pie(expenses,labels=depts,autopct='%1.0f%%',startangle=140,explode=(0, 0.05, 0, 0),shadow=True)
plt.title('Sales Data')
plt.show()
```

# Histogram Example

```python
marks=[45,60,80,55,62,90,89,50,56,60,80,55,62,90,89,58,90,56,67,83]
plt.figure(figsize=(8,5))
plt.hist(marks,bins=5,color='pink',edgecolor='darkblue',alpha=0.8)       
plt.title('Histogram for Student Marks')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.grid(color='grey', linestyle='--', linewidth=0.7)
plt.show()
```
